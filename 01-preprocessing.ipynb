````python
import pandas as pd
````

````python
df = pd.read_csv("./data/interim/german.csv")

# map credit_risk to 0/1 (good/bad)
df["credit_risk"] = df["credit_risk"].map({1: 0, 2: 1})

df.to_csv("./data/processed/german.csv", index=False)
````
